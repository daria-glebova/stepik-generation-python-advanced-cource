# 8. Операции над дробями
Даны две дроби в формате a/b. Напишите программу, которая вычисляет и выводит их сумму, разность, произведение и частное.

#### Формат входных данных
На вход программе подаются две ненулевые дроби, каждая на отдельной строке.

#### Формат выходных данных
Программа должна вывести сумму, разность, произведение и частное двух дробей.

<strong> Примечание. </strong> Обратите внимание на третий тест: исходные дроби сокращать не нужно, а результат нужно.

```python
from fractions import Fraction

a, b = str(input()), str(input())

fraction_a = Fraction(a)
fraction_b = Fraction(b)

print(a, '+', b, '=', fraction_a + fraction_b)
print(a, '-', b, '=', fraction_a - fraction_b)
print(a, '*', b, '=', fraction_a * fraction_b)
print(a, '/', b, '=', fraction_a / fraction_b)
```

# 9. Сумма дробей 1

На вход программе подается натуральное число n. Напишите программу, которая вычисляет и выводит рациональное число, равное значению выражения 

1/1^2 + 1/2^2 + 1/3^2 + 1/4^2 + 1/5^2 + ... + 1/n^2

#### Формат входных данных
На вход программе подается натуральное число n.

#### Формат выходных данных
Программа должна вывести ответ на задачу.

<strong> Примечание. </strong> Результирующая дробь должна быть несократимой.

```python
from fractions import Fraction

n = int(input())

lis = []

for i in range(1, n+1):
    num = Fraction(1, i**2)
    lis.append(num)
    
print(sum(lis))
```

# 10. Сумма дробей 2
На вход программе подается натуральное число n. Напишите программу, которая вычисляет и выводит рациональное число, равное значению выражения 

1/1! + 1/2! + 1/3! + 1/4! + 1/5! + ... + 1/n!

#### Формат входных данных
На вход программе подается натуральное число n.

#### Формат выходных данных
Программа должна вывести ответ на задачу.

<strong> Примечание 1. </strong> Результирующая дробь должна быть несократимой.

<strong> Примечание 2. </strong> Для вычисления факториала можно использовать функцию factorial из модуля math.

```python
from fractions import Fraction
import math

n = int(input())

lis = []

for i in range(1, n+1):
    num = Fraction(1, math.factorial(i))
    lis.append(num)
    
print(sum(lis))
```

Дима учится в седьмом классе и сейчас они проходят обыкновенные дроби с натуральными числителем и знаменателем. Вчера на уроке математики Дима узнал, что дробь называется правильной, если ее числитель меньше знаменателя, и несократимой, если нет равной ей дроби с меньшими натуральными числителем и знаменателем.

Дима очень любит математику, поэтому дома он долго экспериментировал, придумывая и решая разные задачки с правильными несократимыми дробями. Одну из этих задач Дима предлагает решить вам с помощью компьютера.

На вход программе подается натуральное число n. Напишите программу, которая находит наибольшую правильную несократимую дробь с суммой числителя и знаменателя равной n.

#### Формат входных данных
На вход программе подается натуральное число n.

#### Формат выходных данных
Программа должна вывести ответ на задачу.

<strong> Примечание. </strong> Возможно вам потребуется функция gcd(), которая позволяет находить наибольший общий делитель (НОД) двух чисел. Функция реализована в модуле math.

```python
from fractions import Fraction

n = int(input())
f = Fraction(n/2)

if n % 2 == 0:
    if n % 4 == 0:
        k = Fraction(int(f - 1), int(f + 1))
    else:
        k = Fraction(int(f - 2), int(f + 2))
else:
    k = Fraction(int(f - 0.5), int(f + 0.5))
    
print(k)
```

# 12. Упорядоченные дроби
На вход программе подается натуральное число n. Напишите программу, которая выводит в порядке возрастания все несократимые дроби, заключённые между 0 и 1, знаменатель которых не превосходит n.

#### Формат входных данных
На вход программе подается натуральное число n, n > 1.

#### Формат выходных данных
Программа должна вывести ответ на задачу.

<strong> Примечание. </strong> Возможно вам потребуется функция gcd(), которая позволяет находить наибольший общий делитель (НОД) двух чисел. Функция реализована в модуле math.

```python
from fractions import Fraction

n = int(input())
Set = set()

for i in range(1, n+1):
    for j in range(1, n+1):
        num = Fraction(i, j)
        if 0 < num < 1:
            Set.add(num)
        
print(*sorted(Set), sep='\n')
```

# 14. 
Комплексные числа хранятся в списке numbers. Дополните приведенный код так, чтобы он вывел комплексное число с наибольшим модулем и сам модуль числа на отдельных строках.

<strong> Примечание. </strong> Модуль комплексного числа можно вычислить с помощью встроенной функции abs().

```python
numbers = [3 + 4j, 3 + 1j, -7 + 3j, 4 + 8j, -8 + 10j, -3 + 2j, 3 - 2j, -9 + 9j, -1 - 1j, -1 - 10j, -20 + 15j, -21 + 1j, 1j, -3 + 8j, 4 - 6j, 8 + 2j, 2 + 3j]

n = {num: abs(num) for num in numbers}
maximum = max(n, key=abs)
print(maximum, n[maximum], sep='\n')
```

# 15. Сопряженные числа
Дано натуральное число n и два комплексных числа z_1, z_2

Напишите программу, которая вычисляет и выводит значение выражения
z_1^n + z_2^n + \bar{z_1}^n + \bar{z_2}^{n+1}.

#### Формат входных данных
На вход программе подается натуральное число n и два комплексных числа z_1 и z_2, каждое на отдельной строке.

#### Формат выходных данных
Программа должна вывести ответ на задачу.

<strong> Примечание. </strong> Число \bar{z_1} является сопряженным к числу z_1.

```python
n, z_1, z_2 = int(input()), complex(input()), complex(input())

print(z_1**n + z_2**n + (z_1.conjugate())**n + (z_2.conjugate())**(n+1))
```
