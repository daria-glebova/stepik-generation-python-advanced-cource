# 1.

# 3. 
–î–æ–ø–æ–ª–Ω–∏—Ç–µ –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã–π –∫–æ–¥, —á—Ç–æ–±—ã –æ–Ω –≤—ã–≤–µ–ª —Å—É–º–º—É –Ω–∞–∏–±–æ–ª—å—à–µ–π –∏ –Ω–∞–∏–º–µ–Ω—å—à–µ–π —Ü–∏—Ñ—Ä—ã Decimal —á–∏—Å–ª–∞.

```python
from decimal import *

num = Decimal(input())

int_num = int(num)

if int_num == 0:
    print(max(num.as_tuple().digits))
else:
    print(max(num.as_tuple().digits) + min(num.as_tuple().digits))
```

# 4. –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è Decimal —á–∏—Å–ª–æ d. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è:

e^d + ln(d) + lg(d) + sqrt(d)

#### –§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ –¥–µ—Å—è—Ç–∏—á–Ω–æ–µ —á–∏—Å–ª–æ d.

#### –§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –∏—Å–∫–æ–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è.

```python
from decimal import *

d = Decimal(input())

print(d.exp() + d.ln() + d.log10() + d.sqrt())
```

# 8. –û–ø–µ—Ä–∞—Ü–∏–∏ –Ω–∞–¥ –¥—Ä–æ–±—è–º–∏
–î–∞–Ω—ã –¥–≤–µ –¥—Ä–æ–±–∏ –≤ —Ñ–æ—Ä–º–∞—Ç–µ a/b. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç –∏ –≤—ã–≤–æ–¥–∏—Ç –∏—Ö —Å—É–º–º—É, —Ä–∞–∑–Ω–æ—Å—Ç—å, –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –∏ —á–∞—Å—Ç–Ω–æ–µ.

#### –§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–µ –Ω–µ–Ω—É–ª–µ–≤—ã–µ –¥—Ä–æ–±–∏, –∫–∞–∂–¥–∞—è –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

#### –§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Å—É–º–º—É, —Ä–∞–∑–Ω–æ—Å—Ç—å, –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –∏ —á–∞—Å—Ç–Ω–æ–µ –¥–≤—É—Ö –¥—Ä–æ–±–µ–π.

<strong> –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. </strong> –û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ —Ç—Ä–µ—Ç–∏–π —Ç–µ—Å—Ç: –∏—Å—Ö–æ–¥–Ω—ã–µ –¥—Ä–æ–±–∏ —Å–æ–∫—Ä–∞—â–∞—Ç—å –Ω–µ –Ω—É–∂–Ω–æ, –∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω—É–∂–Ω–æ.

```python
from fractions import Fraction

a, b = str(input()), str(input())

fraction_a = Fraction(a)
fraction_b = Fraction(b)

print(a, '+', b, '=', fraction_a + fraction_b)
print(a, '-', b, '=', fraction_a - fraction_b)
print(a, '*', b, '=', fraction_a * fraction_b)
print(a, '/', b, '=', fraction_a / fraction_b)
```

# 9. –°—É–º–º–∞ –¥—Ä–æ–±–µ–π 1

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç –∏ –≤—ã–≤–æ–¥–∏—Ç —Ä–∞—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ, —Ä–∞–≤–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏—é –≤—ã—Ä–∞–∂–µ–Ω–∏—è 

1/1^2 + 1/2^2 + 1/3^2 + 1/4^2 + 1/5^2 + ... + 1/n^2

#### –§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n.

#### –§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ—Ç–≤–µ—Ç –Ω–∞ –∑–∞–¥–∞—á—É.

<strong> –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. </strong> –†–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â–∞—è –¥—Ä–æ–±—å –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –Ω–µ—Å–æ–∫—Ä–∞—Ç–∏–º–æ–π.

```python
from fractions import Fraction

n = int(input())

lis = []

for i in range(1, n+1):
    num = Fraction(1, i**2)
    lis.append(num)
    
print(sum(lis))
```

# 10. –°—É–º–º–∞ –¥—Ä–æ–±–µ–π 2
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç –∏ –≤—ã–≤–æ–¥–∏—Ç —Ä–∞—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ, —Ä–∞–≤–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏—é –≤—ã—Ä–∞–∂–µ–Ω–∏—è 

1/1! + 1/2! + 1/3! + 1/4! + 1/5! + ... + 1/n!

#### –§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n.

#### –§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ—Ç–≤–µ—Ç –Ω–∞ –∑–∞–¥–∞—á—É.

<strong> –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 1. </strong> –†–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â–∞—è –¥—Ä–æ–±—å –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –Ω–µ—Å–æ–∫—Ä–∞—Ç–∏–º–æ–π.

<strong> –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 2. </strong> –î–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª–∞ –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é factorial –∏–∑ –º–æ–¥—É–ª—è math.

```python
from fractions import Fraction
import math

n = int(input())

lis = []

for i in range(1, n+1):
    num = Fraction(1, math.factorial(i))
    lis.append(num)
    
print(sum(lis))
```

–î–∏–º–∞ —É—á–∏—Ç—Å—è –≤ —Å–µ–¥—å–º–æ–º –∫–ª–∞—Å—Å–µ –∏ —Å–µ–π—á–∞—Å –æ–Ω–∏ –ø—Ä–æ—Ö–æ–¥—è—Ç –æ–±—ã–∫–Ω–æ–≤–µ–Ω–Ω—ã–µ –¥—Ä–æ–±–∏ —Å –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã–º–∏ —á–∏—Å–ª–∏—Ç–µ–ª–µ–º –∏ –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª–µ–º. –í—á–µ—Ä–∞ –Ω–∞ —É—Ä–æ–∫–µ –º–∞—Ç–µ–º–∞—Ç–∏–∫–∏ –î–∏–º–∞ —É–∑–Ω–∞–ª, —á—Ç–æ –¥—Ä–æ–±—å –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –ø—Ä–∞–≤–∏–ª—å–Ω–æ–π, –µ—Å–ª–∏ –µ–µ —á–∏—Å–ª–∏—Ç–µ–ª—å –º–µ–Ω—å—à–µ –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—è, –∏ –Ω–µ—Å–æ–∫—Ä–∞—Ç–∏–º–æ–π, –µ—Å–ª–∏ –Ω–µ—Ç —Ä–∞–≤–Ω–æ–π –µ–π –¥—Ä–æ–±–∏ —Å –º–µ–Ω—å—à–∏–º–∏ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã–º–∏ —á–∏—Å–ª–∏—Ç–µ–ª–µ–º –∏ –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª–µ–º.

–î–∏–º–∞ –æ—á–µ–Ω—å –ª—é–±–∏—Ç –º–∞—Ç–µ–º–∞—Ç–∏–∫—É, –ø–æ—ç—Ç–æ–º—É –¥–æ–º–∞ –æ–Ω –¥–æ–ª–≥–æ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä–æ–≤–∞–ª, –ø—Ä–∏–¥—É–º—ã–≤–∞—è –∏ —Ä–µ—à–∞—è —Ä–∞–∑–Ω—ã–µ –∑–∞–¥–∞—á–∫–∏ —Å –ø—Ä–∞–≤–∏–ª—å–Ω—ã–º–∏ –Ω–µ—Å–æ–∫—Ä–∞—Ç–∏–º—ã–º–∏ –¥—Ä–æ–±—è–º–∏. –û–¥–Ω—É –∏–∑ —ç—Ç–∏—Ö –∑–∞–¥–∞—á –î–∏–º–∞ –ø—Ä–µ–¥–ª–∞–≥–∞–µ—Ç —Ä–µ—à–∏—Ç—å –≤–∞–º —Å –ø–æ–º–æ—â—å—é –∫–æ–º–ø—å—é—Ç–µ—Ä–∞.

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –Ω–∞–∏–±–æ–ª—å—à—É—é –ø—Ä–∞–≤–∏–ª—å–Ω—É—é –Ω–µ—Å–æ–∫—Ä–∞—Ç–∏–º—É—é –¥—Ä–æ–±—å —Å —Å—É–º–º–æ–π —á–∏—Å–ª–∏—Ç–µ–ª—è –∏ –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—è —Ä–∞–≤–Ω–æ–π n.

#### –§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n.

#### –§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ—Ç–≤–µ—Ç –Ω–∞ –∑–∞–¥–∞—á—É.

<strong> –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. </strong> –í–æ–∑–º–æ–∂–Ω–æ –≤–∞–º –ø–æ—Ç—Ä–µ–±—É–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏—è gcd(), –∫–æ—Ç–æ—Ä–∞—è –ø–æ–∑–≤–æ–ª—è–µ—Ç –Ω–∞—Ö–æ–¥–∏—Ç—å –Ω–∞–∏–±–æ–ª—å—à–∏–π –æ–±—â–∏–π –¥–µ–ª–∏—Ç–µ–ª—å (–ù–û–î) –¥–≤—É—Ö —á–∏—Å–µ–ª. –§—É–Ω–∫—Ü–∏—è —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω–∞ –≤ –º–æ–¥—É–ª–µ math.

```python
from fractions import Fraction

n = int(input())
f = Fraction(n/2)

if n % 2 == 0:
    if n % 4 == 0:
        k = Fraction(int(f - 1), int(f + 1))
    else:
        k = Fraction(int(f - 2), int(f + 2))
else:
    k = Fraction(int(f - 0.5), int(f + 0.5))
    
print(k)
```

# 12. –£–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã–µ –¥—Ä–æ–±–∏
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è –≤—Å–µ –Ω–µ—Å–æ–∫—Ä–∞—Ç–∏–º—ã–µ –¥—Ä–æ–±–∏, –∑–∞–∫–ª—é—á—ë–Ω–Ω—ã–µ –º–µ–∂–¥—É 0 –∏ 1, –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—å –∫–æ—Ç–æ—Ä—ã—Ö –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç n.

#### –§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n, n > 1.

#### –§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ—Ç–≤–µ—Ç –Ω–∞ –∑–∞–¥–∞—á—É.

<strong> –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. </strong> –í–æ–∑–º–æ–∂–Ω–æ –≤–∞–º –ø–æ—Ç—Ä–µ–±—É–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏—è gcd(), –∫–æ—Ç–æ—Ä–∞—è –ø–æ–∑–≤–æ–ª—è–µ—Ç –Ω–∞—Ö–æ–¥–∏—Ç—å –Ω–∞–∏–±–æ–ª—å—à–∏–π –æ–±—â–∏–π –¥–µ–ª–∏—Ç–µ–ª—å (–ù–û–î) –¥–≤—É—Ö —á–∏—Å–µ–ª. –§—É–Ω–∫—Ü–∏—è —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω–∞ –≤ –º–æ–¥—É–ª–µ math.

```python
from fractions import Fraction

n = int(input())
Set = set()

for i in range(1, n+1):
    for j in range(1, n+1):
        num = Fraction(i, j)
        if 0 < num < 1:
            Set.add(num)
        
print(*sorted(Set), sep='\n')
```

# 13. –û–ø–µ—Ä–∞—Ü–∏–∏ –Ω–∞–¥ –∫–æ–º–ø–ª–µ–∫—Å–Ω—ã–º–∏ —á–∏—Å–ª–∞–º–∏
–î–∞–Ω—ã –¥–≤–∞ –∫–æ–º–ø–ª–µ–∫—Å–Ω—ã—Ö —á–∏—Å–ª–∞. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç –∏ –≤—ã–≤–æ–¥–∏—Ç –∏—Ö —Å—É–º–º—É, —Ä–∞–∑–Ω–æ—Å—Ç—å –∏ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ.

#### –§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –¥–≤–∞ –∫–æ–º–ø–ª–µ–∫—Å–Ω—ã—Ö —á–∏—Å–ª–∞, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

#### –§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ—Ç–≤–µ—Ç –Ω–∞ –∑–∞–¥–∞—á—É.

```python
n, m = complex(input()), complex(input())

print(str(n), '+', str(m), '=', n+m)
print(str(n), '-', str(m), '=', n-m)
print(str(n), '*', str(m), '=', n*m)
```

# 14. 
–ö–æ–º–ø–ª–µ–∫—Å–Ω—ã–µ —á–∏—Å–ª–∞ —Ö—Ä–∞–Ω—è—Ç—Å—è –≤ —Å–ø–∏—Å–∫–µ numbers. –î–æ–ø–æ–ª–Ω–∏—Ç–µ –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã–π –∫–æ–¥ —Ç–∞–∫, —á—Ç–æ–±—ã –æ–Ω –≤—ã–≤–µ–ª –∫–æ–º–ø–ª–µ–∫—Å–Ω–æ–µ —á–∏—Å–ª–æ —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º –º–æ–¥—É–ª–µ–º –∏ —Å–∞–º –º–æ–¥—É–ª—å —á–∏—Å–ª–∞ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω—ã—Ö —Å—Ç—Ä–æ–∫–∞—Ö.

<strong> –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. </strong> –ú–æ–¥—É–ª—å –∫–æ–º–ø–ª–µ–∫—Å–Ω–æ–≥–æ —á–∏—Å–ª–∞ –º–æ–∂–Ω–æ –≤—ã—á–∏—Å–ª–∏—Ç—å —Å –ø–æ–º–æ—â—å—é –≤—Å—Ç—Ä–æ–µ–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ abs().

```python
numbers = [3 + 4j, 3 + 1j, -7 + 3j, 4 + 8j, -8 + 10j, -3 + 2j, 3 - 2j, -9 + 9j, -1 - 1j, -1 - 10j, -20 + 15j, -21 + 1j, 1j, -3 + 8j, 4 - 6j, 8 + 2j, 2 + 3j]

n = {num: abs(num) for num in numbers}
maximum = max(n, key=abs)
print(maximum, n[maximum], sep='\n')
```

# 15. –°–æ–ø—Ä—è–∂–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞
–î–∞–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n –∏ –¥–≤–∞ –∫–æ–º–ø–ª–µ–∫—Å–Ω—ã—Ö —á–∏—Å–ª–∞ z_1, z_2

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç –∏ –≤—ã–≤–æ–¥–∏—Ç –∑–Ω–∞—á–µ–Ω–∏–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è
z_1^n + z_2^n + \bar{z_1}^n + \bar{z_2}^{n+1}.

#### –§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n –∏ –¥–≤–∞ –∫–æ–º–ø–ª–µ–∫—Å–Ω—ã—Ö —á–∏—Å–ª–∞ z_1 –∏ z_2, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

#### –§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ—Ç–≤–µ—Ç –Ω–∞ –∑–∞–¥–∞—á—É.

<strong> –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. </strong> –ß–∏—Å–ª–æ \bar{z_1} —è–≤–ª—è–µ—Ç—Å—è —Å–æ–ø—Ä—è–∂–µ–Ω–Ω—ã–º –∫ —á–∏—Å–ª—É z_1.

```python
n, z_1, z_2 = int(input()), complex(input()), complex(input())

print(z_1**n + z_2**n + (z_1.conjugate())**n + (z_2.conjugate())**(n+1))
```
